```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

def get_dowJones():
    url = 'https://www.dogsofthedow.com/dow-jones-industrial-average-companies.htm'
    headers = {'User-Agent': 'Mozilla/5.0'}
    request = requests.get(url, headers=headers)
    soup = bs(request.text, "html.parser")  # Use the built-in HTML parser
    stats = soup.find('table', class_='tablepress tablepress-id-42 tablepress-responsive')
    pulled_df = pd.read_html(str(stats))[0]
    return pulled_df

def get_sp500():
    url = 'https://www.slickcharts.com/sp500'
    headers = {'User-Agent': 'Mozilla/5.0'}
    request = requests.get(url, headers=headers)
    soup = bs(request.text, "html.parser")  # Use the built-in HTML parser
    stats = soup.find('table', class_='table table-hover table-borderless table-sm')
    df = pd.read_html(str(stats))[0]
    df['% Chg'] = df['% Chg'].str.strip('()-%')
    df['% Chg'] = pd.to_numeric(df['% Chg'])
    df['Chg'] = pd.to_numeric(df['Chg'])
    return df

import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

def convert_market_cap_string(market_cap_str):
    # Check for million (M) or billion (B) and convert to numbers
    if market_cap_str.endswith('M'):
        return float(market_cap_str.strip('M')) * 1e6
    elif market_cap_str.endswith('B'):
        return float(market_cap_str.strip('B')) * 1e9
    else:
        return pd.NA  # Handle other cases (e.g., unknown suffix)

def get_qqq():
    dfs = []  # Create an empty list to store DataFrames

    urls = [
        'https://www.dividendmax.com/market-index-constituents/nasdaq-100',
        'https://www.dividendmax.com/market-index-constituents/nasdaq-100?page=2',
        'https://www.dividendmax.com/market-index-constituents/nasdaq-100?page=3'
    ]

    for url in urls:
        headers = {'User-Agent': 'Mozilla/5.0'}
        request = requests.get(url, headers=headers)
        soup = bs(request.text, "html.parser")  # Use the built-in HTML parser
        stats = soup.find('table', class_='mdc-data-table__table')
        temp = pd.read_html(str(stats))[0]
        dfs.append(temp)  # Append each DataFrame to the list

    df = pd.concat(dfs, ignore_index=True)  # Concatenate all DataFrames in the list
    df.rename(columns={'Market Cap': 'Market Cap $bn'}, inplace=True)

    # Apply the conversion function to the 'Market Cap $bn' column
    df['Market Cap $bn'] = df['Market Cap $bn'].apply(convert_market_cap_string)

    df = df.sort_values('Market Cap $bn', ascending=False)
    return df

# get_dowJones()
get_sp500()
# get_qqq()
```
